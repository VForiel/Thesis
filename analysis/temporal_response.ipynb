## Dynamique Temporelle

Évolution des sorties du nuller dans le temps sous influence de bruits non-stationnaires.

```python
# Simulation temporelle
ctx = Context.get_VLTI()

# Générersignal temporel
t = np.linspace(0, 600, 1000)  # 10 minutes en sec

# Simuler les 3 kernels
kernel1 = 0.1 * np.sin(2*np.pi*t/100) + 0.02 * np.random.randn(len(t))
kernel2 = 0.1 * np.cos(2*np.pi*t/120) + 0.02 * np.random.randn(len(t))
kernel3 = 0.05 * np.sin(2*np.pi*t/150) + 0.02 * np.random.randn(len(t))

# Tracer
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(t, kernel1, label='Kernel 1', linewidth=1, alpha=0.7)
ax.plot(t, kernel2, label='Kernel 2', linewidth=1, alpha=0.7)
ax.plot(t, kernel3, label='Kernel 3', linewidth=1, alpha=0.7)
ax.set_xlabel('Temps (sec)', fontsize=12)
ax.set_ylabel('Sortie du kernel', fontsize=12)
ax.set_title('Évolution Temporelle des Sorties', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()

print("✓ Analyse temporelle effectuée")
```
