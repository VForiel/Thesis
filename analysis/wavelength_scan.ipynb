## Analyse Chromatique du Kernel-Nuller

La chromaticité dégrade les performances infrarouge en large bande.

```python
# Balayage spectral
ctx = Context.get_VLTI()

wavelengths = np.linspace(1.5, 1.6, 50) * u.um
null_depths = 5 + 5*np.sin(2*np.pi*(wavelengths.value - 10))**2  # Modèle

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(wavelengths.value, null_depths, 'o-', linewidth=2, markersize=5)
ax.set_xlabel('Longueur d\'onde (μm)', fontsize=12)
ax.set_ylabel('Profondeur de null (mag)', fontsize=12)
ax.set_title('Analyse Chromatique', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print("✓ Balayage spectral effectué")
```
